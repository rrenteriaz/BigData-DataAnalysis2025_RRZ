```python
# ==============================================================
# Hoja de respuestas - Limpieza y visualización con Pandas
# ==============================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --------------------------------------------------------------
# 1. Importación de datos (Titanic - datos reales)
# --------------------------------------------------------------
url_titanic_dirty = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df_salary = pd.read_csv(url_titanic_dirty)

# Ejercicio 1.1: Mostrar las primeras 5 filas
df_salary.head(5)

# Ejercicio 1.2: Mostrar el tamaño del dataset (filas, columnas)
df_salary.shape

# Ejercicio 1.3: Ver información general (tipos de datos y nulos)
df_salary.info()

# --------------------------------------------------------------
# 2. Filtrado básico de datos
# --------------------------------------------------------------
# Ejercicio 2.1: Filtrar columnas 'Name', 'Fare' y 'Age'
df_subset = df_salary[['Name', 'Fare', 'Age']]

# Ejercicio 2.2: Filtrar filas con Fare mayor a 50
df_high_salary = df_subset[df_subset['Fare'] > 50]

# --------------------------------------------------------------
# 3. Manejo de valores nulos
# --------------------------------------------------------------
# Ejercicio 3.1: Contar valores nulos por columna
df_salary.isnull().sum()

# Ejercicio 3.2: Eliminar filas con nulos en 'Fare'
df_no_null_salary = df_salary.dropna(subset=['Fare'])

# Ejercicio 3.3: Rellenar nulos en 'Age' con 0
df_salary['Age'] = df_salary['Age'].fillna(0)

# --------------------------------------------------------------
# 4. Manejo de duplicados
# --------------------------------------------------------------
# Ejercicio 4.1: Contar filas duplicadas
df_salary.duplicated().sum()

# Ejercicio 4.2: Eliminar duplicados en 'Name'
df_salary = df_salary.drop_duplicates(subset=['Name'])

# --------------------------------------------------------------
# 5. Limpieza de texto
# --------------------------------------------------------------
# Ejercicio 5.1: Convertir 'Name' a mayúsculas
df_salary['Name'] = df_salary['Name'].str.upper()

# Ejercicio 5.2: Eliminar espacios en blanco
df_salary['Name'] = df_salary['Name'].str.strip()

# --------------------------------------------------------------
# 6. Ejercicio con otro dataset real: InsideAirbnb
# --------------------------------------------------------------
url_airbnb = "http://data.insideairbnb.com/united-states/ny/new-york-city/2024-12-04/visualisations/listings.csv"
df_airbnb = pd.read_csv(url_airbnb)

# Ejercicio 6.1: Limpiar columna 'price'
df_airbnb['price'] = df_airbnb['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Ejercicio 6.2: Filtrar precios > 50
df_airbnb = df_airbnb[df_airbnb['price'] > 50]

# Ejercicio 6.3: Manejar fechas en 'last_review'
df_airbnb['last_review'] = pd.to_datetime(df_airbnb['last_review'], errors='coerce')

# --------------------------------------------------------------
# 7. Visualización básica
# --------------------------------------------------------------
# Ejercicio 7.1: Histograma de tarifas (Fare)
df_salary['Fare'].hist(bins=20)
plt.title("Distribución de tarifas")
plt.xlabel("Fare")
plt.ylabel("Cantidad de pasajeros")
plt.show()

# Ejercicio 7.2: Gráfico de barras - Top 10 barrios en Airbnb
df_airbnb['neighbourhood'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 barrios con más alojamientos")
plt.ylabel("Número de alojamientos")
plt.show()

# Ejercicio 7.3: Dispersión Edad vs Tarifa
sns.scatterplot(data=df_salary, x='Age', y='Fare')
plt.title("Tarifa vs Edad")
plt.xlabel("Edad")
plt.ylabel("Fare")
plt.show()
```
