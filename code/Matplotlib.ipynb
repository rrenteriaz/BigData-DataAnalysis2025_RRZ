# HUE plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load data
df = sns.load_dataset('iris')

# Scatter plot: sepal vs petal
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species', palette='Set1', s=100, marker="d")
plt.title("Iris Species by Sepal and Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.grid(True, alpha=0.3)
#plt.show()
plt.savefig("iris_scatter.png", dpi=300)
# Save as pDF
plt.savefig("iris_scatter.pdf")
#plt.show()
```

```python
# Scatter plot: sepal vs petal
plt.figure(figsize=(8, 6))
sns.lineplot(data=df, x='sepal_length', y='petal_length', hue='species', palette='Set1')
plt.title("Iris Species by Sepal and Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.grid(True, alpha=0.3)
plt.show()
```

# Line plot

```python
url = "https://raw.githubusercontent.com/owid/covid-19-data/master/scripts/input/un/population_latest.csv"
df_pop = pd.read_csv(url)

# Filter top 5 countries
top_countries = ['India', 'China', 'United States', 'Indonesia', 'Pakistan']
df_filtered = df_pop[df_pop['entity'].isin(top_countries)]

# Simulated time series (use 2000â€“2020 with fake growth)
years = np.arange(2000, 2021)
plt.figure(figsize=(12, 6))
for _, row in df_filtered.iterrows():
    pop_2020 = row['population']
    growth_rate = np.random.uniform(0.008, 0.015)
    pop_over_time = [pop_2020 / (1 + growth_rate)**(2020 - year) for year in years]
    plt.plot(years, pop_over_time, label=row['entity'], linewidth=2)

plt.title("Simulated Population Growth (2000â€“2020)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

# Annotate

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(42)

# Generate synthetic app ratings (realistic distribution)
# Most apps cluster around 4.0â€“4.8, some low ratings
ratings = np.concatenate([
    np.random.normal(4.5, 0.5, 800),  # Majority: good apps
    np.random.normal(2.5, 0.8, 200)   # Some poorly rated apps
])

# Clip to valid range [1.0, 5.0]
ratings = np.clip(ratings, 1.0, 5.0)

# Create DataFrame
df = pd.DataFrame({'Rating': ratings})

# Plot
plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'], kde=True, bins=30, color='teal', alpha=0.8, edgecolor='white')
plt.title("Distribution of App User Ratings", fontsize=14, fontweight='bold')
plt.xlabel("User Rating (1â€“5)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.axvline(df['Rating'].mean(), color='red', linestyle='--', linewidth=2,
            label=f'Mean Rating: {df["Rating"].mean():.2f}')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

# Bubble plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed
np.random.seed(123)

# List of countries and continents
countries = [
    ('Japan', 'Asia'), ('Germany', 'Europe'), ('Nigeria', 'Africa'), ('Brazil', 'Americas'),
    ('India', 'Asia'), ('USA', 'Americas'), ('France', 'Europe'), ('Ethiopia', 'Africa'),
    ('China', 'Asia'), ('Canada', 'Americas'), ('UK', 'Europe'), ('Egypt', 'Africa'),
    ('Mexico', 'Americas'), ('Vietnam', 'Asia'), ('Kenya', 'Africa')
]

# Generate synthetic data
data = []
for country, continent in countries:
    # Base life expectancy by continent (realistic)
    base_life = {
        'Asia': 75, 'Europe': 80, 'Americas': 76, 'Africa': 60
    }[continent]

    # Base GDP
    base_gdp = {
        'Asia': 40000, 'Europe': 45000, 'Americas': 50000, 'Africa': 8000
    }[continent]

    # Add noise
    life = np.random.normal(base_life, 5)
    gdp = np.random.lognormal(np.log(base_gdp), 0.5)  # log-normal spread
    pop = np.random.lognormal(18, 1.5)  # realistic population spread (1M to 1B)

    data.append({
        'Country': country,
        'Continent': continent,
        'Life Expectancy': max(life, 40),  # cap min
        'GDP per Capita': gdp,
        'Population': pop
    })

df = pd.DataFrame(data)

# Sort by population and take top 10 for clean plot
df_top = df.nlargest(10, 'Population')

# Bubble plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    df_top['GDP per Capita'],
    df_top['Life Expectancy'],
    s=df_top['Population'] / 10_000_000,  # scale size
    c=pd.Categorical(df_top['Continent']).codes,  # color by continent
    cmap='Set1',
    alpha=0.8,
    edgecolors='black',
    linewidth=0.6
)

# Labels
plt.xscale('log')
plt.xlabel("GDP per Capita (USD, log scale)")
plt.ylabel("Life Expectancy (Years)")
plt.title("Synthetic: Life Expectancy vs GDP (Bubble Size = Population)", fontsize=14)

# Color legend
handles, _ = scatter.legend_elements(prop="colors", num=len(df_top['Continent'].unique()))
legend1 = plt.legend(handles, df_top['Continent'].unique(), title="Continent", loc="lower right")

# Add country labels
for i in df_top.iterrows():
    plt.annotate(
        i[1]['Country'],
        (i[1]['GDP per Capita'], i[1]['Life Expectancy']),
        textcoords="offset points",
        xytext=(0, 5),
        ha='center',
        fontsize=9,
        alpha=0.9
    )

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
